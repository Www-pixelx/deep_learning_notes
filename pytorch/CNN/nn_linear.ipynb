```python
import torch
import torchvision
from torch.utils.data import Dataset,DataLoader
from torch import nn
```

```python
dataset = torchvision.datasets.CIFAR10("../data",
                                       train = False,
                                       transform = torchvision.transforms.ToTensor(),
                                       download=True)
dataloader = DataLoader(dataset, batch_size=64,drop_last=True)
```

```python
class Tudui_Linear(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(196608, 10)

    def forward(self, input):
        output = self.linear(input)
        return output
```

```python
tudui = Tudui_Linear()

for data in dataloader:
    imgs, targets = data
    # 把图片展开成一列
    imgs = torch.flatten(imgs) # imgs = imgs.reshape(1,1,1,-1)
    print(imgs.shape)
    output = tudui(imgs)
    print(output.shape)
```
